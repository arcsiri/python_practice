```python
import copy
nums = [-1,0,1,2,-1,-4]
nums.sort()
k=0
l=[]
while k<len(nums):
  first=nums[k]
  duplicate_nums=copy.copy(nums)
  duplicate_nums.remove(first)
  for i in range(len(duplicate_nums)-1):
    second=duplicate_nums[i]
    for j in range(i+1,len(duplicate_nums)):
      third=duplicate_nums[j]
      sum=second+third
      if sum+first==0:
        triplet=sorted([first,second,third])
        if triplet not in l:
          l.append(triplet)
  k=k+1
print(l)
```
